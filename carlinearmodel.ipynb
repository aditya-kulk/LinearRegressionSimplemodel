```python
new_df = df.drop(['Gender', 'Credit Card Debt'], axis=1)
```

```python
new_df.head()
```

```python
X=new_df.drop('Car Purchase Amount',axis=1)
y=new_df['Car Purchase Amount']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
sc=StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)

# Use the same scaler to transform the test data
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
linear_regression_prediction=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r2_score(linear_regression_prediction,y_test)
```

```python
from sklearn.linear_model import Ridge
rg=Ridge()
```

```python
param_grid={'alpha':[0.1,0.5,1,10,100],'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg','saga']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(param_grid=param_grid,estimator=rg,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2sc=r2_score(y_test,y_pred)
print("Mean Squared Error on Test Data:", mse)
print('R2',r2sc)
```
